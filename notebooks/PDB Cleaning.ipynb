```python
from pdbfixer import PDBFixer
from simtk.openmm.app import PDBFile
import os
import multiprocessing as mp
```

```python
DATA_PATH='../data/'
MUTATED_PDBS_PATH=DATA_PATH+'pdbs_mutated/'
CLEANED_PDBS_PATH=DATA_PATH+'pdbs_mutated_cleaned/'
```

```python
fixer = PDBFixer(filename=MUTATED_PDBS_PATH+'1A22_A_B_CA171A.pdb')
fixer.findMissingResidues()
```

```python
fixer.findNonstandardResidues()
fixer.replaceNonstandardResidues()
```

```python
fixer.removeHeterogens(True)
fixer.findMissingAtoms()
fixer.addMissingAtoms()
fixer.addMissingHydrogens(7.0)
```

```python
PDBFile.writeFile(fixer.topology, fixer.positions, \
                  open(CLEANED_PDBS_PATH+'1A22_A_B_CA171A.pdb', 'w'))
```

```python
i=0
for fname in os.listdir(MUTATED_PDBS_PATH):
    print(fname)
    i=i+1
    if i>10:
        break
```

```python
def pdb_clean_gen(fname):
    fixer = PDBFixer(filename=MUTATED_PDBS_PATH+fname)
    fixer.findMissingResidues()
    fixer.findNonstandardResidues()
    fixer.replaceNonstandardResidues()
    fixer.removeHeterogens(True)
    fixer.findMissingAtoms()
    fixer.addMissingAtoms()
    fixer.addMissingHydrogens(7.0)
    
    PDBFile.writeFile(fixer.topology, fixer.positions, \
                      open(CLEANED_PDBS_PATH+fname, 'w'))

p = mp.Pool(5)
for _ in p.imap_unordered(pdb_clean_gen, os.listdir(MUTATED_PDBS_PATH)):
    pass
print('finished')
```
